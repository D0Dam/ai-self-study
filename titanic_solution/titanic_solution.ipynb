### Load dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_df = pd.read_csv("./test.csv")
train_df = pd.read_csv("./train.csv")
```

```python
train_df.head(5)
```

```python
train_df.set_index('PassengerId', inplace=True)
test_df.set_index('PassengerId', inplace=True)
```

```python
train_df.head(5)
```

```python
train_index = train_df.index
test_index = test_df.index
```

```python
y_train_df = train_df.pop("Survived")
```

### Data preprocessing

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
test_df.isnull().sum() / len(test_df) * 100
```

```python
train_df.isnull().sum() / len(train_df) * 100
```

```python
del test_df["Cabin"]
del train_df["Cabin"]
```

```python
all_df = pd.concat([train_df, test_df])
all_df
```

```python
(all_df.isnull().sum() / len(all_df)).plot(kind='bar')
plt.show()
```

```python
del all_df["Name"]
del all_df["Ticket"]
```

```python
all_df
```

```python
all_df["Sex"] = all_df["Sex"].replace({ "male": 0, "female": 1 }).astype(int)
```

```python
all_df
```

```python
all_df["Embarked"].unique()
```

```python
all_df["Embarked"] = all_df["Embarked"].replace({ "S": 0, "C": 1, "Q": 2, np.nan: 99}).astype(int)
```

```python
pd.get_dummies(all_df["Embarked"], prefix="embarked").astype(int)
```

```python
matrix_df = pd.merge(all_df, pd.get_dummies(all_df["Embarked"], prefix="embarked").astype(int), left_index=True, right_index=True)
matrix_df
```

```python
matrix_df.corr()
```

```python
all_df.groupby("Pclass")["Age"].mean()
```

```python
all_df.groupby("Sex")["Age"].mean()
```

```python
all_df.loc[(all_df["Pclass"] ==1) & (all_df["Age"].isnull())]
```

```python
all_df.loc[(all_df["Pclass"] ==1) & (all_df["Age"].isnull()), "Age"] = 39.16
all_df.loc[(all_df["Pclass"] ==2) & (all_df["Age"].isnull()), "Age"] = 29.51
all_df.loc[(all_df["Pclass"] ==3) & (all_df["Age"].isnull()), "Age"] = 24.82
```

```python
all_df.isnull().sum()
```

```python
all_df.groupby("Pclass")["Fare"].mean()
```

```python
all_df[all_df["Fare"].isnull()]
```

```python
all_df.loc[all_df["Fare"].isnull(), "Fare"] = 13.30
```

```python
del all_df["Embarked"]
```

```python
all_df["Pclass"] = all_df["Pclass"].replace({1:"A", 2:"B", 3:"C"})
all_df
```

```python
pd.get_dummies(all_df["Pclass"], prefix="Pclass").astype(int)
```

```python
all_df = pd.merge(all_df, pd.get_dummies(all_df["Pclass"], prefix="Pclass").astype(int), left_index=True, right_index=True)
```

```python
del all_df["Pclass"]
```

```python
all_df
```

```python
all_df = pd.merge(all_df, matrix_df[["embarked_0", "embarked_1", "embarked_2", "embarked_99"]], left_index=True, right_index=True)
all_df
```

```python
train_df = all_df[all_df.index.isin(train_index)]
test_df = all_df[all_df.index.isin(test_index)]
```

```python
train_df.head(5)
```

### Build Model

```python
x_data = train_df.to_numpy()
y_data = y_train_df.to_numpy()
```

```python
from sklearn.linear_model import LogisticRegression
cls = LogisticRegression(max_iter=1000)
cls.fit(x_data, y_data)
```

```python
cls.intercept_
```

```python
cls.predict(test_df.values)
```

```python
test_df.index
```

```python
x_test = test_df.to_numpy()
y_test =cls.predict(x_test)
y_test 
```

```python
result = np.concatenate( (test_index.values.reshape(-1,1), cls.predict(x_test).reshape(-1,1)  ) , axis=1)
```

```python
df_submssion =pd.DataFrame(result, columns=["PassengerId","Survived"])
df_submssion
```

```python
df_submssion.to_csv("submission_result.csv",index=False)
```
